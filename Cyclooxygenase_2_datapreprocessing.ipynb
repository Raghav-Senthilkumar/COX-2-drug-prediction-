```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
targetprotein = target.search('Cyclooxygenase-2')
targets = pd.DataFrame.from_dict(targetprotein)
targets
```

```python
selected_target = targets.target_chembl_id[1]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)

df
```

```python
df2 = df[df.standard_value.notna()]
df2
```

```python
for col in df2.columns:
    print(col)
```

```python
standard_value = df2.standard_value
molecule_chembl_id = df2.molecule_chembl_id
canonical_smiles = df2.canonical_smiles
```

```python
df3 = pd.concat([molecule_chembl_id,canonical_smiles, standard_value], axis=1)
df3
```

```python
df3.isnull().values.any()
```

```python
df3 = df3[df3.standard_value.notna()]
df3 = df3[df3.molecule_chembl_id.notna()]
df3 = df3[df3.canonical_smiles.notna()]
```

```python
df3.isnull().values.any()
```

```python
df3.to_csv('COX-2_data.csv', index=False)
```

The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.
