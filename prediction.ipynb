```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('cox-2_padldescriptors')
```

```python
df2 = pd.read_csv('COX-2_druglikeness-2.csv')
df2
```

```python
pIC50 = df2['pIC50']
nhdonors = df2['Number of H donors']
nhacceptors = df2['Number of H Acceptors']
logP = df2['LogP']
molmass = df2['molecular mass']
PSA = df2['PSA']
numrotbonds = df2['numrotbonds']
df['Number of H donors'] =nhdonors
df['Number of H Acceptors'] = nhacceptors
df['LogP'] = logP
df['molecular mass'] =molmass
df['PSA'] = PSA
df['numrotbonds'] = numrotbonds

```

```python
pIC50 = df2['pIC50']
```

```python
df
```

```python
df.shape
```

```python
pIC50.shape
```

```python
df = df.drop(['Name'], axis=1)
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))
df = selection.fit_transform(df)
```

```python
df.shape
```

```python
df = pd.DataFrame(df)
df
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df, pIC50, test_size=0.05)
```

```python
model = RandomForestRegressor(n_estimators=300)
model.fit(X_train, Y_train)
r2 = model.score(X_test, Y_test)
print("Accuracy of the RandomForestRegressor with 300 trees:" +str(r2))
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.datasets import load_diabetes
est = HistGradientBoostingRegressor().fit(X_train, Y_train)
```

```python
score = est.score(X_test,Y_test)
print("Accuracy of the HistGradientBoostingRegressor: "+ str(score))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(random_state = 0)

# Fitting the Decision Tree Regression model to the data
model.fit(X_train, Y_train)
```

```python
print("Accuracy of the DecisionTreeRegressor: "+ str(model.score(X_test, Y_test)))
```
